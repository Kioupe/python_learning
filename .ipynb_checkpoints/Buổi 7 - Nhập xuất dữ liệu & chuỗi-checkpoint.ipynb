# Nhập xuất dữ liệu và chuỗi (String)

### Xuất dữ liệu

Lệnh `print()` là lệnh giúp python xuất dữ liệu ra màn hình

Cú pháp:

`print(*object, sep=' ', end='\n')`

Trong đó:

- objects – là các giá trị sẽ được xuất ra màn hình
- sep – phân cách giữa các giá trị. Giá trị mặc định của sep là một khoảng trắng.
- end – là kí tự được in ra sau khi các giá trị của objects đã được in. Mặc định của end là ‘\n’ – xuống dòng mới.
##### objects có thể là nhiều giá trị, mỗi giá trị cách nhau một dấu phẩy

Ví dụ:

`print("Hello","mọi người")`

Ở đây, objects là "HEllo" và "mọi người"

##### Đọc thêm tài liệu tại đây: https://www.w3schools.com/python/ref_func_print.asp

Thực hiện một lệnh print với objects có một giá trị là biến:

```python
a="mọi người"
print("Hello",a)
```

Thực hiện một lệnh print có sep là '\t':

```python
a="mọi người"
print("Hello",a,sep="\t")
```

Thực hiện một lệnh print có end là '.':

```python
print("Hello mọi người",end='.')
```

### Nhập dữ liệu

Lệnh `input()` là lệnh giúp python nhận dữ liệu nhập từ bàn phím

Cú pháp:

`input(prompt)`

Trong đó:

- prompt là nội dung in ra màn hình

Giá trị trả về của lệnh `input()` mang kiểu dữ liệu là String (chuỗi)

Thực hiện doạn lệnh yêu cầu nhập vào tên người dùng, sau đó in ra dòng chữ "Xin chào [tên người dùng]":

```python
user=input("Nhập tên người dùng : ") #input sẽ nhận kiểu string
print("Xin chào",user)
```

Thực hiện một đoạn lệnh yêu cầu nhập vào bán kính một đường tròn, sau đó in ra diện tích và chu vi hình tròn đó

```python
a=float(input("nhập bán kính hình tròn: "))
print("Chu vi hình hình tròn: ", 2*3.14*a )
print("Diện tích hình hình tròn: ", 3.14*a**2 )
```

### Chuỗi dữ liệu

Khai báo chuỗi:
    
`chuoi = "hello, mọi người cố gắng nhé!!"`

Trong python, chuỗi có thể được đặt trong hai dấu nháy đơn, hoặc hai dấu nháy kép

### Truy xuất ký tự trong chuỗi

Cú pháp:

`chuỗi[vị trí]`

Vị trí bắt đầu là 0

Vị trí cuối cùng của chuỗi là -1

Ví dụ:

```python
chuoi = "hello mọi người"
print(len(chuoi))
print(chuoi[14])

```

```python
chuoi = input()
print(len(chuoi))
print(chuoi[-10])

```

```python
chuoi = input()
print(len(chuoi))
print(chuoi[5])
```

### Cắt chuỗi

Cú pháp:

`chuỗi[a: b]`

Trong đó
- a là vị trí đầu
- b là vị trí cuối +1


```python
chuoi =input()
chuoi_cut=chuoi[7:-1]#lấy từ 7-9
print(chuoi_cut)
```

```python
#cắt ngược chuỗi
chuoi =input()
chuoi_cut=chuoi[::-1]#cắt ngược chuỗi
print(chuoi_cut)
```

### Toán tử với chuỗi

#### Toán tử +

Toán tử + giúp gắn liền 2 chuỗi:

Ví dụ:

```python
chuoi_a = input()
chuoi_b = input()
chuoi_c = chuoi_a+" "+chuoi_b # ghép chuỗi a vs khoảng trắng rồi ghép chuỗi b
print(chuoi_c)
```

#### Toán tử *

Toán tử * giúp lặp lại đoạn chuỗi

Ví dụ:

```python
chuoi_a = input()
chuoi_b = input()
chuoi_c = chuoi_a*3#chỉ được nhân vs số nguyên
print(chuoi_c)
```

### Một số lệnh xử lý String

#### len – trả về độ dài của chuỗi

Cú pháp:

`len(chuỗi)`

Kết quả trả về có kiểu dữ liệu là int (kiểu số nguyên)

Ví dụ:

```python
chuoi=input()
a=len(chuoi)
for i in range(a):
    print(i,chuoi[i])
    i=i+1
```

### Fun fact

Kiểu dữ liệu chuỗi là một kiểu dữ liệu tốn khá nhiều tài nguyên RAM của máy

1 ký tự được lưu trong chuỗi, chúng ta sẽ cần 1 byte RAM

##### *1KB = 1024 Byte
##### *1MB = 1024 KByte
##### *1GB = 1024 MByte
##### *1TB = 1024 GByte

Hãy lập trình một đoạn, tính toán và in ra 1TB thì bằng:
Bao nhiêu Byte?
Bao nhiêu KB?
Bao nhiêu MB?
Bao nhiêu GB?

```python
print("1 tb=",1064,"gb = ",1064**2,"mb =",1064**3,"kb = ",(1064**3)*1000,"b")
```

Nhập vào một địa chỉ email, kiểm tra xem địa chỉ email đó có đúng cú pháp hay không, và có phải là email được đăng ký từ gmail hay không!

In ra kết quả là True hay False cho từng yêu cầu!

(Email đúng cú pháp là email chỉ có duy nhất 1 ký tự @)

```python
email=input("Nhập địa chỉ email: ")
#chuyển email nhập thành chữ thường
email1=email.lower()
#biến kiểm tra email có nhập đúng hay không
kt=0
#lấy vị trí đuôi của email
vitri=email.find('@')
duoiemail=email[vitri:]
#print(duoiemail)
#kiểm tra xem email được nhập có đúng 1 ký tự @ hay không
if email1.count('@')==1:
    #kiểm tra xem email được nhập có chứa ký tự trắng không
    if email1.count(' ')<1:
        #kiểm tra đuôi email có hợp lệ không
        if email1[-4:]==".com":
            #kiểm tra đuôi email có đúng không và có tên user chưa
            if duoiemail.count(".")<=1 and email1[0:vitri]!='':
                print("In Email: TRUE")
                kt=1
                #kiểm tra xem email có được đăng ký 
                if email1[-10:]=="@gmail.com":
                    print('In Gmail: TRUE.')   
                else:
                    print('in Gmail: FALSE') 
if kt==0:
    print("In Email: FALSE")
    
```

Với số lượng và màu sắc của hình vuông được nhập từ bàn phím

```python
import turtle
t=turtle.Turtle()
turtle.bgcolor("black")
t.pensize(2)
b=input("nhập màu của hình: ")
a=int(input("số lượng hình vuông: "))
#thay đổi màu
t.color(b)
#lặp theo số lượng hình vuông
for i in range(a):
    for i in range(4):
        t.fd(100)
        t.rt(90)    
    t.left(360/a)
turtle.mainloop() 
```
